```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
fmri = sns.load_dataset("fmri")
fmri
```

```python
sns.lineplot(x="timepoint", y="signal", data=fmri,hue="event",markers=True,style="event")
plt.show()
```

```python
sns.set(style="whitegrid")
data = pd.read_csv("breast_cancer.csv")

sns.relplot(data=data, x="569",y="30",hue =data.iloc[:,-1] )
plt.show()
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.relplot(y='tip',x='total_bill',data=tips,hue='smoker',style='sex')
```

```python
flights = sns.load_dataset("flights")
flights
```

```python
sns.lineplot(data=flights,x="year",y="passengers",hue="month")
```

```python
flights_w = flights.pivot(columns="year",index="month",values="passengers")
flights_w
```

```python
sns.relplot(data=flights_w,kind='line')
```

```python
sns.catplot(data=flights_w.transpose(),kind='box')
```
